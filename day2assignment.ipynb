```
pos={"x":0, "y":0}
n=int(input()) 
for i in range(n) :
    move = input().split(" ") 
    if move[0].lower() == "up":
      pos["y"] += int(move[1]) 
    elif move[0].lower() == "down":
      pos["y"] -= int(move[1]) 
    elif move[0].lower() == "left":
      pos["x"] -= int(move[1]) 
    elif move[0].lower() == "right":
      pos["x"] += int(move[1]) 
print(int(round((pos["x"]**2+pos["y"]**2)**0.5)))
```
